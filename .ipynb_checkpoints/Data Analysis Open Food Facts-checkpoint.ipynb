## Data cleaning

Since Data is too large (333824 rows & 162 columns) so let's look at how much data is NULL or NaN in each column so that we can get rid of columns which are majorly NULLs

```python
plt.figure(figsize=(5, 20))
df.isnull().mean(axis=0).plot.barh()
plt.title("Proportion of NaNs in each column")
```

```python
# This will give us number of NULLs or NaN in each columns
df.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
# If majorty of data is missing than interpolation will be highly skewed so ,
# Let's drop the columns where 70% or more are NULL
# 70% of 333824 = 233676
df=df.dropna(axis=1,thresh=233676)
df.shape
```

Above result shows that only 29 columns have data in large quantity and all others were majorly missing.

But intuitively, we know that all these above nutrients are not different for each product so we can't interpolate it with some preditive model. Best way to do would to fill in mean/median/mode value for these columns.

```python
### Outlier Detection through Boxplots
sns.boxplot(df.sugars_100g)
```

```python
sns.boxplot(df.proteins_100g)
```

```python
# sugars_100g and proteins_100g have negative values which are obviously not legit so let's remove them 
df.loc[df['sugars_100g'] < 0, 'sugars_100g'] = 0
df.loc[df['proteins_100g'] < 0, 'proteins_100g'] = 0
```

Now we'll remove outliers from these columns. We know 99% data is in 3sigma band (as per quartiles) so lets remove data out present outside this 3quartile range.

```python
# Similarly other columns have outliers, so Removing outliers
df = df[np.abs(df.energy_100g-df.energy_100g.mean())<=(3*df.energy_100g.std())]
df = df[np.abs(df.fat_100g-df.fat_100g.mean())<=(3*df.fat_100g.std())]
df = df[np.abs(df['saturated-fat_100g']-df['saturated-fat_100g'].mean())<=(3*df['saturated-fat_100g'].std())]
df = df[np.abs(df.carbohydrates_100g-df.carbohydrates_100g.mean())<=(3*df.carbohydrates_100g.std())]
df = df[np.abs(df.sugars_100g-df.sugars_100g.mean())<=(3*df.sugars_100g.std())]
df = df[np.abs(df.proteins_100g-df.proteins_100g.mean())<=(3*df.proteins_100g.std())]
df = df[np.abs(df.salt_100g-df.salt_100g.mean())<=(3*df.salt_100g.std())]
df = df[np.abs(df.sodium_100g-df.sodium_100g.mean())<=(3*df.sodium_100g.std())]
```

### Filling missing value

```python
# Filling Mean values where NaN is present
df['energy_100g'].fillna(df['energy_100g'].mean(), inplace=True)
df['fat_100g'].fillna(df['fat_100g'].mean(), inplace=True)
df['saturated-fat_100g'].fillna(df['saturated-fat_100g'].mean(), inplace=True)
df['carbohydrates_100g'].fillna(df['carbohydrates_100g'].mean(), inplace=True)
df['sugars_100g'].fillna(df['sugars_100g'].mean(), inplace=True)
df['proteins_100g'].fillna(df['proteins_100g'].mean(), inplace=True)
df['salt_100g'].fillna(df['salt_100g'].mean(), inplace=True)
df['sodium_100g'].fillna(df['sodium_100g'].mean(), inplace=True)
```

### Let's study Countries & Product ingredients

```python
def splitDataFrameList(df, target_column, separator):
    def splitListToRows(row, row_accumulator, target_column, separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)
    new_rows = []
    df.apply(splitListToRows,axis=1,args = (new_rows,target_column,separator))
    new_df = pd.DataFrame(new_rows)
    return new_df
```

```python
food_countries = splitDataFrameList(df, "countries_en", ",")
countries = food_countries["countries_en"].value_counts()
```

```python
plt.title("Top Food Producing countries")
countries[:10][::-1].plot.barh()
```

```python
food_ingredients = splitDataFrameList(df, "ingredients_text", ",")
ingredients = food_ingredients["ingredients_text"].value_counts()
```

```python
plt.title("Top ingredients in food products")
ingredients[:20][::-1].plot.barh()
```

How much sugar do we eat?

```python
df.countries = df.countries_en.str.lower()
    
def mean(l):
    return float(sum(l)) / len(l)

world_sugars = df[df.sugars_100g.notnull()]

def return_sugars(country):
    return world_sugars[world_sugars.countries == country].sugars_100g.tolist()
    
# Get list of sugars per 100g for some countries
fr_sugars = return_sugars('france') + return_sugars('en:fr')
za_sugars = return_sugars('south africa')
uk_sugars = return_sugars('united kingdom') + return_sugars('en:gb')
us_sugars = return_sugars('united states') + return_sugars('en:us') + return_sugars('us')
sp_sugars = return_sugars('spain') + return_sugars('españa') + return_sugars('en:es')
nd_sugars = return_sugars('netherlands') + return_sugars('holland')
au_sugars = return_sugars('australia') + return_sugars('en:au')
cn_sugars = return_sugars('canada') + return_sugars('en:cn')
de_sugars = return_sugars('germany')

countries = ['FR', 'ZA', 'UK', 'US', 'ES', 'ND', 'AU', 'CN', 'DE']
sugars_l = [mean(fr_sugars), 
            mean(za_sugars), 
            mean(uk_sugars), 
            mean(us_sugars), 
            mean(sp_sugars), 
            mean(nd_sugars),
            mean(au_sugars),
            mean(cn_sugars),
            mean(de_sugars)]
            
y_pos = np.arange(len(countries))
    
plt.bar(y_pos, sugars_l, align='center', alpha=0.5)
plt.title('Average total sugar content per 100g')
plt.xticks(y_pos, countries)
plt.ylabel('Sugar/100g')
    
plt.show()
```
